```python
import librosa
from pydub import AudioSegment
import os
from IPython.display import Audio, display
from elevenlabs.client import ElevenLabs

client = ElevenLabs(
  api_key="sk_8e7b121b6bc20df50f4219b84f354b6e58f82b494266a86b"
)

def speech(text, voice_id, model_id="eleven_multilingual_v2", file_name="output_audio.wav"):
    audio = client.text_to_speech.convert(text=text, voice_id=voice_id, model_id=model_id)
    audio_bytes = b"".join(audio)
    # Guardar el archivo de audio
    with open(file_name, "wb") as f:
        f.write(audio_bytes)
    return file_name

def analyze_music_beats(music_file):
    music, sr = librosa.load(music_file, sr=None)
    tempo, beats = librosa.beat.beat_track(y=music, sr=sr)
    beat_times = librosa.frames_to_time(beats, sr=sr)
    return beat_times

def split_text_into_phrases(text, num_phrases):
    if num_phrases <= 0:  # Handle cases where no beats are detected
        return [text]
    words = text.split()
    avg_words_per_phrase = len(words) // num_phrases
    # Handle the potential ZeroDivisionError if no beats are found
    if avg_words_per_phrase == 0:
        return [text]  # Return the entire text as a single phrase
    phrases = [' '.join(words[i:i + avg_words_per_phrase]) for i in range(0, len(words), avg_words_per_phrase)]
    return phrases

def combine_audio_with_music(phrases, beat_times, music_file, output_file):
    music = AudioSegment.from_file(music_file)
    combined = AudioSegment.silent(duration=len(music))

    for i, phrase in enumerate(phrases):
        phrase_audio_file = f'phrase_{i}.mp3'
        speech(phrase, voice_id="9oPKasc15pfAbMr7N6Gs", file_name=phrase_audio_file)
        phrase_audio = AudioSegment.from_file(phrase_audio_file)

        # Encuentra el tiempo de inicio del beat correspondiente
        if i < len(beat_times):
            start_time = beat_times[i] * 1000  # Convertir a milisegundos
        else:
            start_time = len(combined)  # Si se acaban los beats, añadir al final

        combined = combined.overlay(phrase_audio, position=start_time)
        os.remove(phrase_audio_file)  # Limpia archivos temporales

    combined.export(output_file, format='mp3')
    return output_file

def main():
    # Define el texto que deseas convertir a voz
    prompt = """¿Ni la mejor computadora?
¿Lo dice usted que es un simple humano?
Se muestra tan ufano
Pero le llegó la hora
Porque su ego lo devora
Y se cree superior a ultranza
Pero con una muestra alcanza
Hoy las bombas que crearon sus mentes
Son más inteligentes
Que los idiotas que las lanzan"""

    # Análisis de los beats de la música
    music_file = '/content/BSG-BEAT_HIP-HOP-1.mp3'  # Asegúrate de tener este archivo
    beat_times = analyze_music_beats(music_file)

    # Divide el texto en frases
    num_phrases = len(beat_times)
    phrases = split_text_into_phrases(prompt, num_phrases)

    # Combina la respuesta de voz con la base musical
    output_file = 'output_with_music.mp3'
    combined_file = combine_audio_with_music(phrases, beat_times, music_file, output_file)

    # Reproduce el archivo combinado
    display(Audio(combined_file, autoplay=True))

if __name__ == "__main__":
    main()
```
