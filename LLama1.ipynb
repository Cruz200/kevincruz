```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Función para generar texto con un prompt específico para rap
def generate_rap_response(prompt, max_length=300):
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids
    output_ids = model.generate(
        input_ids,
        max_length=max_length,
        temperature=0.7,
        top_k=50,
        top_p=0.9,
        num_return_sequences=1
        repetition_penalty=1.2
    )
    output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return output_text
```

```python

# Contexto inicial para generar una respuesta coherente
context = "tus respuestas seran siempre seran simulando una cancion rap y seran en base a lo que yo te diga:\n"

# Texto de entrada inicial
input_text = "¿quien rapea mejor tu o yo, yo soy una campeon y tu solo una calculadora?"

# Combina el contexto y el texto de entrada
prompt = context + input_text

# Genera la respuesta orientada a crear una canción de rap
response = generate_rap_response(prompt)




# Actualiza el contexto con la respuesta generada
context += input_text + "\n" + response + "\n"

print(prompt)
```
